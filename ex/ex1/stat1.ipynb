# מבוא לסטטיסטיקה והסתברות לכיתות ז' ו-ח'

## 1. הקדמה לסטטיסטיקה

### מהי סטטיסטיקה ולמה היא חשובה?
סטטיסטיקה היא תחום במדע שמאפשר לנו לאסוף, לנתח, לפרש ולהציג נתונים. היא חשובה מכיוון שהיא עוזרת לנו להבין ולתאר את העולם סביבנו באמצעות מספרים.

### איסוף נתונים
איסוף נתונים יכול להיעשות בדרכים שונות, כמו סקרים ומדגמים.

## 2. ארגון וייצוג נתונים

### טבלאות שכיחות
טבלה שמציגה את מספר הפעמים שכל ערך מופיע בנתונים.

### גרפים
נלמד על יצירת גרפים עמודיים, גרפים עוגתיים ותרשימים ב-Python באמצעות matplotlib.

```python
import matplotlib.pyplot as plt

# נתונים לדוגמה
data = [4, 8, 15, 16, 23, 42, 8, 15, 15, 16, 23, 42]

# טבלת שכיחות
frequency = {x: data.count(x) for x in set(data)}

# גרף עמודים
plt.bar(frequency.keys(), frequency.values())
plt.xlabel('ערך')
plt.ylabel('שכיחות')
plt.title('גרף עמודים של נתונים')
plt.show()

# גרף עוגה
plt.pie(frequency.values(), labels=frequency.keys(), autopct='%1.1f%%')
plt.title('גרף עוגה של נתונים')
plt.show()

# ## 3. מדדי מרכז

# ### ממוצע
# הממוצע הוא סכום כל הערכים חלקי מספר הערכים.
average = sum(data) / len(data)
print(f'הממוצע הוא: {average}')

# ### חציון
# החציון הוא הערך האמצעי כאשר הערכים מסודרים בסדר עולה.
sorted_data = sorted(data)
median = sorted_data[len(sorted_data) // 2]
print(f'החציון הוא: {median}')

# ### שכיח
# השכיח הוא הערך שמופיע הכי הרבה פעמים.
mode = max(set(data), key=data.count)
print(f'השכיח הוא: {mode}')

# ## 4. מדדי פיזור

# ### טווח
# הטווח הוא ההפרש בין הערך הגבוה ביותר לערך הנמוך ביותר.
range_value = max(data) - min(data)
print(f'הטווח הוא: {range_value}')
```

```python
# ## 5. הקדמה להסתברות

# ### ניסוי הסתברותי
# נציג דוגמאות של ניסויים הסתברותיים כמו הטלת מטבע והטלת קוביה.

import random

# הטלת מטבע
def coin_toss():
    return 'עץ' if random.randint(0, 1) == 0 else 'פלי'

# הטלת קוביה
def dice_roll():
    return random.randint(1, 6)
```

```python
results = [coin_toss() for _ in range(100)]
print(f'תוצאות הטלת מטבע: {results.count("עץ")} עץ, {results.count("פלי")} פלי')
```

```python
results = [dice_roll() for _ in range(100)]
print(f'תוצאות הטלת קוביה: {[results.count(i) for i in range(1, 7)]}')
```

```python
# הסתברות לקבל מספר זוגי בהטלת קוביה
p_dice = 3 / 6
print(f'הסתברות לקבל מספר זוגי בהטלת קוביה: {p_dice}')
```
