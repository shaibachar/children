## load libs

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

## Generate random data following a normal distribution

```python
np.random.seed(0)  # For reproducibility
mean = 0
std_dev = 1
data = np.random.normal(mean, std_dev, 1000)
```

```python
len(data)
```

## Plot the histogram of the data

```python
plt.figure(figsize=(12, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue', label='Histogram of Data')
```

## Step 3: Overlay the probability density function (PDF)

```python
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution PDF')
```

## Adding mean and standard deviation lines

```python
plt.axvline(mean, color='red', linestyle='-', linewidth=2, label=f'Mean: {mean}')
plt.axvline(mean + std_dev, color='green', linestyle='--', linewidth=2, label=f'Mean + 1 Std Dev: {mean + std_dev}')
plt.axvline(mean - std_dev, color='green', linestyle='--', linewidth=2, label=f'Mean - 1 Std Dev: {mean - std_dev}')
```

## Adding labels and legend

```python
plt.title('Normal Distribution with Mean and Standard Deviation')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
```

## Display the plot

```python
plt.show()
```

## Explanation

```python
explanation = """
Normal Distribution:
A normal distribution, also known as a Gaussian distribution, is a continuous probability distribution that is symmetrical around its mean. It follows the '68-95-99.7' rule, meaning:

- About 68% of the data falls within one standard deviation (std) from the mean.
- About 95% of the data falls within two standard deviations from the mean.
- About 99.7% of the data falls within three standard deviations from the mean.

In the plot:
- The blue bars represent the histogram of the data, showing the frequency of data points.
- The black line represents the normal distribution's probability density function (PDF).
- The red line represents the mean of the data.
- The green dashed lines represent one standard deviation above and below the mean.

This visual helps to understand how the data is distributed and how much of the data falls within certain ranges around the mean.
"""

print(explanation)
```

## another explenation 

```python


# Define the parameters of the normal distribution
mean = 0
std_dev = 2

# Generate a range of x values
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Calculate the PDF values for these x values
pdf = stats.norm.pdf(x, mean, std_dev)

# Plot the PDF
plt.figure(figsize=(12, 6))
plt.plot(x, pdf, 'b-', label='Normal Distribution PDF')
plt.fill_between(x, pdf, alpha=0.3, color='blue')

# Adding labels and title
plt.title('Probability Density Function of a Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

# Display the plot
plt.show()

# Explanation
explanation = """
Probability Density Function (PDF):
The PDF of a normal distribution describes how the values of a random variable are distributed. The mean () indicates the center of the distribution, and the standard deviation () measures the spread.

- The peak of the curve represents the most probable value.
- The area under the curve within one standard deviation (1) from the mean covers about 68% of the data.
- The area within two standard deviations (2) covers about 95% of the data.
- The area within three standard deviations (3) covers about 99.7% of the data.

The PDF is useful for understanding the likelihood of different outcomes and is a key tool in statistical analysis.
"""

print(explanation)
```
