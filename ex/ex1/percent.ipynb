```python
# # מבוא לאחוזים

# ## מהו אחוז?
# אחוז הוא דרך לבטא חלק מתוך שלם, והוא מיוצג על ידי הסימן %. אחוז אחד הוא חלק אחד מתוך מאה.
# לדוגמה, אם יש לנו 100 תפוחים ו-25 מהם הם אדומים, נוכל לומר ש-25% מהתפוחים הם אדומים.

# ## חישוב אחוזים

# ### חישוב אחוז מתוך מספר
# נניח שיש לנו מספר מסוים ואנחנו רוצים לדעת כמה אחוזים ממנו הם מספר אחר.
# לדוגמה, אם יש לנו 200 תלמידים בבית הספר ו-50 מהם הם תלמידים חדשים, נוכל לחשב מה האחוז של התלמידים החדשים.

# מספר התלמידים הכולל
total_students = 200

# מספר התלמידים החדשים
new_students = 50

# חישוב האחוז
percent_new_students = (new_students / total_students) * 100
print(f'אחוז התלמידים החדשים הוא: {percent_new_students}%')

# ### חישוב מספר מתוך אחוז
# נניח שאנחנו יודעים את האחוז ורוצים לדעת כמה הוא מתוך מספר מסוים.
# לדוגמה, אם אנחנו יודעים ש-30% מהתלמידים הם תלמידים מצטיינים ורוצים לדעת כמה תלמידים מצטיינים יש אם יש 200 תלמידים.

# אחוז התלמידים המצטיינים
percent_excellent_students = 30

# חישוב מספר התלמידים המצטיינים
excellent_students = (percent_excellent_students / 100) * total_students
print(f'מספר התלמידים המצטיינים הוא: {excellent_students}')

# ### העלאה באחוזים
# נניח שמחיר מוצר הוא 150 ש"ח ואנחנו רוצים להעלות את המחיר ב-10%.
# נוכל לחשב את המחיר החדש לאחר ההעלאה.

# מחיר המוצר המקורי
original_price = 150

# אחוז ההעלאה
percent_increase = 10

# חישוב מחיר ההעלאה
increase_amount = (percent_increase / 100) * original_price

# חישוב המחיר החדש
new_price = original_price + increase_amount
print(f'המחיר החדש לאחר העלאה של 10% הוא: {new_price} ש"ח')

# ### הורדה באחוזים
# נניח שמחיר מוצר הוא 200 ש"ח ואנחנו רוצים להוריד את המחיר ב-15%.
# נוכל לחשב את המחיר החדש לאחר ההורדה.

# מחיר המוצר המקורי
original_price = 200

# אחוז ההורדה
percent_decrease = 15

# חישוב מחיר ההורדה
decrease_amount = (percent_decrease / 100) * original_price

# חישוב המחיר החדש
new_price = original_price - decrease_amount
print(f'המחיר החדש לאחר הורדה של 15% הוא: {new_price} ש"ח')

# ### תרגול
# חישבו את האחוז של מספר מתוך מספר אחר, למשל: כמה אחוזים הם 45 מתוך 150.

# המספרים
part = 45
whole = 150

# חישוב האחוז
percent = (part / whole) * 100
print(f'45 הוא {percent}% מתוך 150')

# חישבו מספר מתוך אחוז, למשל: כמה הם 25% מתוך 80.

# האחוז והמספר
percent = 25
whole = 80

# חישוב המספר
part = (percent / 100) * whole
print(f'25% מתוך 80 הוא: {part}')
```

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Function to calculate percentage and plot
def calculate_and_plot(total, part, operation, percent):
    fig, ax = plt.subplots()
    
    if operation == 'Calculate Percentage':
        # Calculate percentage
        result = (part / total) * 100
        ax.bar(['Total', 'Part'], [total, part], color=['blue', 'orange'])
        ax.set_title(f'{part} is {result:.2f}% of {total}')
    elif operation == 'Calculate Part':
        # Calculate part from percentage
        result = (percent / 100) * total
        ax.bar(['Total', 'Part'], [total, result], color=['blue', 'green'])
        ax.set_title(f'{percent}% of {total} is {result:.2f}')
    elif operation == 'Increase by Percentage':
        # Calculate increase by percentage
        increase = (percent / 100) * total
        new_value = total + increase
        ax.bar(['Original', 'Increase', 'New Value'], [total, increase, new_value], color=['blue', 'red', 'purple'])
        ax.set_title(f'{total} increased by {percent}% is {new_value:.2f}')
    elif operation == 'Decrease by Percentage':
        # Calculate decrease by percentage
        decrease = (percent / 100) * total
        new_value = total - decrease
        ax.bar(['Original', 'Decrease', 'New Value'], [total, decrease, new_value], color=['blue', 'red', 'purple'])
        ax.set_title(f'{total} decreased by {percent}% is {new_value:.2f}')
    
    plt.show()

# Widgets for user input
total_input = widgets.IntSlider(value=100, min=1, max=1000, step=1, description='Total:')
part_input = widgets.IntSlider(value=50, min=1, max=1000, step=1, description='Part:')
percent_input = widgets.FloatSlider(value=10, min=1, max=100, step=0.1, description='Percent:')
operation_selector = widgets.RadioButtons(
    options=['Calculate Percentage', 'Calculate Part', 'Increase by Percentage', 'Decrease by Percentage'],
    description='Operation:',
    disabled=False
)

# Interactive function
ui = widgets.VBox([total_input, part_input, percent_input, operation_selector])
out = widgets.interactive_output(calculate_and_plot, {'total': total_input, 'part': part_input, 'operation': operation_selector, 'percent': percent_input})

display(ui, out)
```
