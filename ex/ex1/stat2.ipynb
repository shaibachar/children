```python
# # בעיות מילוליות בסטטיסטיקה והסתברות לכיתות ז' ו-ח' עם פתרונות

# ## בעיה 1: ממוצע ציונים
# יש לכם את ציוניהם של 5 תלמידים במבחן: 85, 90, 78, 92, ו-88.
# חשבו את הממוצע של הציונים.

# נתוני הציונים
grades = [85, 90, 78, 92, 88]

# חישוב הממוצע
average_grade = sum(grades) / len(grades)
print(f'הממוצע של הציונים הוא: {average_grade}')

# ## בעיה 2: חציון גילאים
# גילאים של קבוצת אנשים הם: 15, 18, 21, 24, 27.
# חשבו את החציון של הגילאים.

# נתוני הגילאים
ages = [15, 18, 21, 24, 27]

# חישוב החציון
sorted_ages = sorted(ages)
median_age = sorted_ages[len(sorted_ages) // 2]
print(f'החציון של הגילאים הוא: {median_age}')

# ## בעיה 3: שכיח צבעים
# בסקר שנערך, נשאלו 10 אנשים על הצבע האהוב עליהם. התשובות היו:
# אדום, כחול, ירוק, אדום, כחול, כחול, ירוק, אדום, כחול, צהוב.
# מצאו את הצבע השכיח ביותר.

# נתוני הצבעים
colors = ['אדום', 'כחול', 'ירוק', 'אדום', 'כחול', 'כחול', 'ירוק', 'אדום', 'כחול', 'צהוב']

# חישוב השכיח
mode_color = max(set(colors), key=colors.count)
print(f'הצבע השכיח ביותר הוא: {mode_color}')

# ## בעיה 4: טווח טמפרטורות
# טמפרטורות שנמדדו במהלך השבוע הן: 20, 22, 19, 24, 21, 23, 20.
# חשבו את הטווח של הטמפרטורות.

# נתוני הטמפרטורות
temperatures = [20, 22, 19, 24, 21, 23, 20]

# חישוב הטווח
range_temperature = max(temperatures) - min(temperatures)
print(f'הטווח של הטמפרטורות הוא: {range_temperature}')

# ## בעיה 5: הסתברות בהטלת מטבע
# מהי ההסתברות לקבל 'עץ' בהטלת מטבע הוגן?

# מספר המקרים המוצלחים (עץ)
successful_outcomes = 1

# מספר המקרים האפשריים (עץ או פלי)
total_outcomes = 2

# חישוב ההסתברות
p_coin = successful_outcomes / total_outcomes
print(f'ההסתברות לקבל "עץ" בהטלת מטבע היא: {p_coin}')

# ## בעיה 6: הסתברות בהטלת קוביה
# מהי ההסתברות לקבל מספר זוגי בהטלת קוביה?

# מספר המקרים המוצלחים (2, 4, 6)
successful_outcomes = 3

# מספר המקרים האפשריים (1 עד 6)
total_outcomes = 6

# חישוב ההסתברות
p_dice = successful_outcomes / total_outcomes
print(f'ההסתברות לקבל מספר זוגי בהטלת קוביה היא: {p_dice}')

# ## בעיה 7: ממוצע גילאים בקבוצה
# בקבוצה של 6 תלמידים הגילאים הם: 13, 14, 13, 15, 14, 13.
# חשבו את ממוצע הגילאים.

# נתוני הגילאים
ages_group = [13, 14, 13, 15, 14, 13]

# חישוב הממוצע
average_age_group = sum(ages_group) / len(ages_group)
print(f'ממוצע הגילאים בקבוצה הוא: {average_age_group}')

# ## בעיה 8: חציון ציונים בכיתה
# הציונים של תלמידים בכיתה הם: 75, 85, 95, 65, 80, 70, 90.
# חשבו את החציון של הציונים.

# נתוני הציונים
scores = [75, 85, 95, 65, 80, 70, 90]

# חישוב החציון
sorted_scores = sorted(scores)
median_score = sorted_scores[len(sorted_scores) // 2]
print(f'החציון של הציונים הוא: {median_score}')

# ## בעיה 9: שכיח שמות בכיתה
# שמות תלמידים בכיתה הם: אורי, דני, מיכל, יעל, אורי, דני, דני.
# מצאו את השם השכיח ביותר.

# נתוני השמות
names = ['אורי', 'דני', 'מיכל', 'יעל', 'אורי', 'דני', 'דני']

# חישוב השכיח
mode_name = max(set(names), key=names.count)
print(f'השם השכיח ביותר הוא: {mode_name}')

# ## בעיה 10: טווח גבהים בקבוצה
# גבהים של קבוצה הם: 150, 160, 170, 155, 165, 160.
# חשבו את הטווח של הגבהים.

# נתוני הגבהים
heights = [150, 160, 170, 155, 165, 160]

# חישוב הטווח
range_height = max(heights) - min(heights)
print(f'הטווח של הגבהים הוא: {range_height}')
```
